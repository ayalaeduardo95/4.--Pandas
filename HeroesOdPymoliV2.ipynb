```python
# Imports
import pandas as pd
import numpy as np
# Loading the csv file
file_csv = "Resources/Gamedata.csv"
game_df = pd.read_csv(file_csv)
game_df.head()
```

## Player Count

* Display the total number of players


```python
# Total number of players
Total_Players = len(game_df["SN"].unique())
# Creating a Data Frame for the total number of players
Players = pd.DataFrame({"Total Players":Total_Players}, index = [0])
Players
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
# Purchasing Analysis
# Number of Unique Items
Total_Items = len(game_df["Item ID"].unique())
# Average Price
Average_Price = game_df["Price"].mean()
# Total Number of Purchases
Total_Purchases = len(game_df["Purchase ID"].unique())
# Total Revenue
Total_Revenue = game_df["Price"].sum()
# Construct new dataframe
Purchasing_Analysis = pd.DataFrame ({"Number of Unique Items" : [Total_Items],
                                     "Average Price" : [Average_Price],
                                     "Number of Purchases" : [Total_Purchases],
                                     "Total Revenue" : [Total_Revenue]})
# Give format to columns "${:,.2f}
Purchasing_Analysis["Average Price"] = Purchasing_Analysis["Average Price"].map("${:,.2f}".format)
Purchasing_Analysis["Total Revenue"] = Purchasing_Analysis["Total Revenue"].map("${:,.2f}".format)
# Dataframe
Purchasing_Analysis
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python
# Gender Demographics
# Removing Duplicate Players
Total_Users = game_df[["SN","Age", "Gender","Price"]]
Total_Users_WD = Total_Users.drop_duplicates(["SN"])
# Percentage and Count of Male Players
Total_Genders = Total_Users_WD["Gender"].value_counts()
Percentage_Genders = Total_Genders / Total_Players
# Create Dataframe for calculations
Gender_Demographics = pd.DataFrame({"Total Count" : Total_Genders,
                                    "Percentage of Players" : Percentage_Genders})
# Give format to columns {:,.2%}
Gender_Demographics["Percentage of Players"] = Gender_Demographics["Percentage of Players"].map("{:,.2%}".format)
Gender_Demographics
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Top Spenders
# Run a new group by to have all values including duplicates
User_ID_Grouped = game_df.groupby("SN")
# Calculations
Purchase_Count_SN = User_ID_Grouped["Price"].count()
Average_Purchase_SN = User_ID_Grouped["Price"].mean()
Total_Purchase_SN = User_ID_Grouped["Price"].sum()
# Create DF
Top_Spenders = pd.DataFrame({"Purchase Count" : Purchase_Count_SN, 
                             "Average Purchase Price" : Average_Purchase_SN,
                             "Total Purchase Value" : Total_Purchase_SN})
# Format 
Top_Spenders["Average Purchase Price"] =Top_Spenders["Average Purchase Price"].map("${:,.2f}".format)
Top_Spenders["Total Purchase Value"] =Top_Spenders["Total Purchase Value"].map("${:,.2f}".format)
# Sort
Top_Spenders = Top_Spenders.sort_values("Purchase Count", ascending = False)
# Head to see top 5
Top_Spenders.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# Most Popular Items
Most_Popular = game_df.loc[:,["Item ID","Item Name","Price"]]
# Grouping Data by Item ID & Item Name
Most_Popular_Grouped = game_df.groupby(["Item ID","Item Name"])
# Calculations for DataFrame
Purchase_Count_Items = Most_Popular_Grouped["Price"].count()
Avg_Purchase_Items = Most_Popular_Grouped["Price"].mean()
Total_Purchase_Items = Most_Popular_Grouped["Price"].sum()
# Create DF
Most_Popular_Items = pd.DataFrame({"Purchase Count" : Purchase_Count_Items,
                                   "Item Price" : Avg_Purchase_Items,
                                   "Total Purchase Value" : Total_Purchase_Items})
# Format
Most_Popular_Items["Item Price"] =Most_Popular_Items["Item Price"].map("${:,.2f}".format)
Most_Popular_Items["Total Purchase Value"] =Most_Popular_Items["Total Purchase Value"].map("${:,.2f}".format)
# Sort
Most_Popular_Items = Most_Popular_Items.sort_values("Purchase Count", ascending = False)
# Run
Most_Popular_Items.head()
```
